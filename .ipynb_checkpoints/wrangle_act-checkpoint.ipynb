## Project Motivation keys

### Key Points
Key points to keep in mind when data wrangling for this project:

- **You only want original ratings (no retweets) that have images**. Though there are 5000+ tweets in the dataset, not all are dog ratings and some are retweets.
- Assessing and cleaning the entire dataset completely would require a lot of time, and is not necessary to practice and demonstrate your skills in data wrangling. Therefore, the requirements of this project are only to assess and clean at least `8 quality` issues and at least `2 tidiness` issues in this dataset.
- Cleaning includes merging individual pieces of data according to the rules of tidy data.
- The fact that the rating numerators are greater than the denominators does not need to be cleaned. This unique rating system is a big part of the popularity of WeRateDogs.
- You do not need to gather the tweets beyond August 1st, 2017. You can, but note that you won't be able to gather the image predictions for these tweets since you don't have access to the algorithm used.

```python
count = 0
fails_dict = {}
start = timer()
# Save each tweet's returned JSON as a new line in a .txt file
with open('tweet_json.txt', 'w') as outfile:
    for tweet_id in tweet_ids:
        count += 1
        print(str(count) + ": " + str(tweet_id))
        try:
            tweet = api.get_status(tweet_id, tweet_mode='extended')
            print("Success")
            json.dump(tweet._json, outfile)
            outfile.write('\n')
        except tweepy.TweepError as e:
            print("Fail")
            fails_dict[tweet_id] = e
            pass
end = timer()
print(end - start)
print(fails_dict)
```

### Quality

**`df_all_tweets` table**

- Tweets in JSON files have typically 27 objects, however, `df_all_tweets` has 32 columns. is the data loading done correctly? what are the new data and their values? Do they really exist in the source file?

- `id` and `id_str` coloumns don't match. Tweeter developer guide for JSON objects suggest `id_str` is the correct user ID while `id` is not. 
- `display_text_range` is stored as a list and is not hashable. It makes it very difficult to use some of the pandas methods on the dataframe. Chnage it to 2 columns with start and finish as integers for futur operation(maybe in ML).
- `display_textt-range` is a list. there are 23 lines with first item in the list being non-zero. it means that the texts is truncated. once mergerd, instead of relying on `text` column from `df_enhanced`, we should use `full_text` from `df_all_tweets`

- Tweeter api couldn't download 22 of the tweets probably because those tweets are deleted. Find those and remove from df_enhanced
- non-null values in the following columns suggest the tweet was written in reply to another user and is not original. We need to remove them. 
['in_reply_to_screen_name','in_reply_to_status_id',
'in_reply_to_status_id_str','in_reply_to_user_id','in_reply_to_user_id_str']
- `source` column contains an html like tag format. 3 different sources of data is clear. they need to be parsed.
- `created_at` is an string object, it must be in timestamp format.


**`df_enhanced` table**

- `id_str` is an object made of strings. Considering `id-str` is a categorical data, all of them must be string.
- There are 23 data lines with `rating_denominator` different than 10. It doesn't follow their rating system. These data must be further looked at after merging and possibly corrected/dropped.
- `timestamp` is an string object, it must be in timestamp format.

**`df_images` table**

- `id_str` is an object made of strings. Considering `id-str` is a categorical data, all of them must be string.


### Tidiness

- ['in_reply_to_status_id', 'in_reply_to_user_id', 'source'] columns are duplicated between `df_all_tweets` and `df_enhanced`. drop them from `df_enhanced` and keep `df_all_tweets`.
- `df_enhanced` and `df_all_tweets` should be merged into one table based on their id_str. `df_all_tweets` is downloaded recently so when there is an overlap between columns/data, the data from `df_all_tweets` should be used.


**`df_all_tweets` table**

- following columns have no datapoints. 
['contributors','coordinates','geo','place']

**`df_enhanced` table**
- all 'None' and 'NaN' values are not Pandas's null values.
- all of these 4 categorical columns, ['doggo','floofer','pupper','puppo'], can be combined to one categorical column.
